```python
# 1mb = 1024 kb

def size_process(item):
  if str(item)[-1]=="M":
    res=float(str(item).replace("M",""))
    res=res*1024
    return res
  elif str(item)[-1]=="k":
    res=float(str(item).replace("k",""))
    return res
  else:
    return str(np.NaN)
```

```python
# Function to extract the lowest version as an integer
def extract_version(version):
    if pd.isna(version) or version == "Varies with device":
        return 0  # Replace missing or varying values with 0

    version = str(version)  # Convert to string to avoid AttributeError
    version = version.split(' ')[0]  # Take only the first part before 'and up' or '-'
    version = version.split('-')[0]  # If range, take the lower bound
    version = version.replace('W', '')  # Remove any non-numeric characters like 'W'

    try:
        return int(float(version))  # Convert to integer
    except ValueError:
        return 0  # If conversion fails, return 0

# Apply the function to clean data
df_copy['Android Ver'] = df_copy['Android Ver'].apply(extract_version)

# Display unique cleaned values
print(df_copy['Android Ver'].unique())
```
